# 02. Aggregations & Joins
- Grouping
    - Count, sum, avg, min, max
- Windowing
    - Row_number, rank, dense_rank
- Joins
    - Inner, outer (left, right, full), semi, anti, cross

```python
from pyspark.sql import Row

# Sample data for DataFrame 1
data1 = [
    Row(id=1, value='A'),
    Row(id=2, value='B'),
    Row(id=3, value='C'),
    Row(id=4, value='D')
]

# Sample data for DataFrame 2
data2 = [
    Row(id=3, desc='foo'),
    Row(id=4, desc='bar'),
    Row(id=5, desc='baz'),
    Row(id=6, desc='qux')
]

df1 = spark.createDataFrame(data1)
df2 = spark.createDataFrame(data2)

```

```python
display(df1)
display(df2)
```

## Joins

```python
df_inner_join = df1.join(df2, on="id", how="inner")
display(df_inner_join)
```

```python
df_left_join = df1.join(df2, on="id", how="left_outer")
display(df_left_join)
```

```python
df_right_join = df1.join(df2, on="id", how="right_outer")
display(df_right_join)
```

```python
df_full_join = df1.join(df2, on="id", how="full_outer")
display(df_full_join)
```

```python
df_anti_join = df1.join(df2, on="id", how="anti")
display(df_anti_join)
```

```python
df_anti_join = df2.join(df1, on="id", how="anti")
display(df_anti_join)
```

```python
df_semi_join = df1.join(df2, on="id", how="semi")
display(df_semi_join)
```

## Grouping

```python
from pyspark.sql import Row

# Sample data for grouping and windowing
data = [
    Row(dept="Sales", name="Alice", salary=5000),
    Row(dept="Sales", name="Bob", salary=4800),
    Row(dept="Sales", name="Charlie", salary=5200),
    Row(dept="Sales", name="Sham", salary=5200),
    Row(dept="Sales", name="Priya", salary=5400),
    Row(dept="HR", name="David", salary=4000),
    Row(dept="HR", name="Mike", salary=4100),
    Row(dept="HR", name="Amy", salary=4100),
    Row(dept="HR", name="Eve", salary=4200),
    Row(dept="IT", name="Frank", salary=6000),
    Row(dept="IT", name="Grace", salary=6100),
    Row(dept="IT", name="Heidi", salary=5900)
]

df = spark.createDataFrame(data)
display(df)
```

```python
from pyspark.sql.functions import sum,avg,count,min,max

df_grouped = df.groupBy("dept")\
    .agg(avg("salary"))\
    .withColumnRenamed("avg(salary)", "avg_salary")

display(df_grouped)
```

```python
from pyspark.sql.functions import sum,avg,count,min,max

df_grouped = df.groupBy("dept").agg(
    avg("salary").alias("avg_salary"),
    sum("salary").alias("total_cost_of_salary"),
    count("salary").alias("headcount"),
    min("salary").alias("min_salary"),
    max("salary").alias("max_salary")
)

display(df_grouped)
```

## Windowing

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, dense_rank, percent_rank, ntile, lag, lead, cume_dist, sum, avg

w = Window.orderBy("salary")

df_windowed = df.withColumn("row_number", row_number().over(w)) \
    .withColumn("rank", rank().over(w)) \
    .withColumn("dense_rank", dense_rank().over(w)) \
    .withColumn("percent_rank", percent_rank().over(w)) \
    .withColumn("ntile_2", ntile(2).over(w)) \
    .withColumn("lag_salary", lag("salary", 1).over(w)) \
    .withColumn("lead_salary", lead("salary", 1).over(w)) \
    .withColumn("cume_dist", cume_dist().over(w)) 

display(df_windowed)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, dense_rank, percent_rank, ntile, lag, lead, cume_dist, sum, avg

w = Window.partitionBy("dept").orderBy("salary")

df_windowed = df.withColumn("row_number", row_number().over(w)) \
    .withColumn("rank", rank().over(w)) \
    .withColumn("dense_rank", dense_rank().over(w)) \
    .withColumn("percent_rank", percent_rank().over(w)) \
    .withColumn("ntile_2", ntile(2).over(w)) \
    .withColumn("lag_salary", lag("salary", 1).over(w)) \
    .withColumn("lead_salary", lead("salary", 1).over(w)) \
    .withColumn("cume_dist", cume_dist().over(w)) 

display(df_windowed)
```
